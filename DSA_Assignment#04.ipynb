## Outlier Detection and Handling

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Creating a box plot for the 'Fare','SibSp' and 'Parch' column

num_cols = ['Fare','SibSp','Parch']
for i in num_cols:
    plt.figure()
    plt.boxplot(data[i])
    plt.title(i)
```

The columns 'SibSp' and 'Parch' represent the number of siblings/spouses and parents/children aboard the Titanic, respectively. 

'SibSp' and 'Parch' columns will be considered normal as they have a distribution that closely resembles a normal distribution.

In case of 'Fare', they have lot of outliers, so removing them

```python

Q1 = data['Fare'].quantile(0.25)
Q3 = data['Fare'].quantile(0.75)

IQR = Q3 - Q1


low_bound = Q1 - 1.5 * IQR
up_bound = Q3 + 1.5 * IQR

outliers = data[(data['Fare'] < low_bound) | (data['Fare'] > up_bound)]
```

```python
outlier_indices = outliers.index

outlier_indices
```

```python
outliers
```

```python
# Removing the outliers from the DataFrame

data.drop(outlier_indices, inplace=True)
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler

#Selecting only numeric columns

numeric_features = data.select_dtypes(include=['int64', 'float64'])

#Separate the target variable 'Survived' from the numeric feature set

target = numeric_features['Survived']
numeric_features = numeric_features.drop('Survived', axis=1)

#Initialize the MinMaxScaler

scaler = MinMaxScaler()

#Fit and transform the numeric feature set

scaled_features = scaler.fit_transform(numeric_features)
```

```python
#Creating a new DataFrame with the scaled features

scaled_data = pd.DataFrame(scaled_features, columns=numeric_features.columns)
```

```python
scaled_data.describe()
```

```python
#Adding the 'Survived' column back to the scaled data

scaled_data['Survived'] = target

scaled_data
```
