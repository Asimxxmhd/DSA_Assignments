# Reading the dataset to pyhton environment

```python
import pandas as pd
```

```python
data = pd.read_csv('iris.csv')
```

# Display the columns in the dataset

```python
columns = data.columns
print("Columns in the dataset:", columns)
```

#  Mean of each column of the dataset

```python
column_means = data.mean()
print("Mean of each column:\n",column_means)
```

# Checking for the null values present in the dataset

```python
null_counts = data.isnull().sum()
print("Null value counts:\n",null_counts)
```

# Meaningful visualizations using the dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(data['SL'], bins=20)
plt.xlabel('Sepal Length (SL)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
plt.scatter(data['SL'], data['SW'])
plt.xlabel('Sepal Length (SL)')
plt.ylabel('Sepal Width (SW)')
plt.title('Scatter Plot: SL vs SW')
plt.show()
```

```python
plt.scatter(data['PL'], data['PW'])
plt.xlabel('Petal Length (PL)')
plt.ylabel('Petal Width (PW)')
plt.title('Scatter Plot: PL vs PW')
plt.show()
```

```python
sns.boxplot(x='Classification', y='PL', data=data)
plt.xlabel('Classification')
plt.ylabel('Petal Length (PL)')
plt.title('Box Plot of Petal Length by Classification')
plt.xticks(rotation=45)
plt.show()
```
